```python
one_hot = pd.get_dummies(df[['room_type_reserved', 'market_segment_type', 'type_of_meal_plan']])
label_encoder = LabelEncoder()
df['room_type_encoded'] = label_encoder.fit_transform(df['room_type_reserved'])
df['market_segment_encoded'] = label_encoder.fit_transform(df['market_segment_type'])
df['meal_plan_encoded'] = label_encoder.fit_transform(df['type_of_meal_plan'])
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
df[['lead_time', 'avg_price_per_room', 'no_of_week_nights', 'no_of_weekend_nights']] = scaler.fit_transform(
    df[['lead_time', 'avg_price_per_room', 'no_of_week_nights', 'no_of_weekend_nights']]
)
normalizer = MinMaxScaler()
df[['lead_time', 'avg_price_per_room', 'no_of_week_nights', 'no_of_weekend_nights']] = normalizer.fit_transform(
    df[['lead_time', 'avg_price_per_room', 'no_of_week_nights', 'no_of_weekend_nights']]
)
```

```python
df.info()
```

```python
df['booking_status'] = df['booking_status'].apply(lambda x: 1 if x == 'Cancelled' else 0)
```

```python
# IQR Method
Q1 = df['lead_time'].quantile(0.25)
Q3 = df['lead_time'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['lead_time'] < (Q1 - 1.5 * IQR)) | (df['lead_time'] > (Q3 + 1.5 * IQR))]
print(outliers)
df.head(10)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.select_dtypes(include=['float64', 'int64']))

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
print('Explained variance ratio:', pca.explained_variance_ratio_)
```

```python
market_segment_group = df.groupby('market_segment_type').agg({
    'booking_status': 'mean',  # Calculate cancellation rate
    'avg_price_per_room': 'mean',  # Average price per room
    'lead_time': 'mean'  # Average lead time
}).reset_index()
print(market_segment_group)
```

```python
room_type_group = df.groupby('room_type_reserved').agg({
    'booking_status': 'mean',
    'avg_price_per_room': 'mean',
    'lead_time': 'mean'
}).reset_index()
print(room_type_group)
```

```python
arrival_month_group = df.groupby('arrival_month').agg({
    'booking_status': 'mean',
    'avg_price_per_room': 'mean',
    'lead_time': 'mean'
}).reset_index()
print(arrival_month_group)
```

```python
sns.barplot(x='market_segment_type', y='avg_price_per_room', data=market_segment_group, palette='magma')
plt.title('Average Room Price by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Average Price Per Room')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x='room_type_reserved', y='lead_time', data=room_type_group, palette='Blues_d')
plt.title('Average Lead Time by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Lead Time (Days)')
plt.xticks(rotation=45)
plt.show()
```
